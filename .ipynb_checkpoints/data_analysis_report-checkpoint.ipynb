```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../diamonds_train.csv')

'''
Project's target: analyse the way the variables given in the dataset affect to the price's variation in order to decide how to treat them in the model
we'll build for Module 3 Project.
'''

data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
'''
Looking at below's numerical variables correlation, we see a high corr between carat (weight) and x, y, z. In order to avoid multicollinearity in the
model I won't consider them.

There is a high correlation between carat and price, which makes sense: you pay more for a bigger diamond. Also correlation between price and depth and
table is very low, so first thoughts are to not include them in the model.
'''

data.corr()
```

```python
'''
Once analysed the correlation between all the numerical values, I've got to the conclusion to include in the model, at first, the parameter carat.

Now, I want to see price's descriptive statistics grouping by each parameter (cut, color and clarity). For that I'm going to plot boxes to make it more visual, otherwise it
would be too much information and too many variables.

I'll use price_per_carat (=price/carat) variable to avoid diamond's size biasing.
'''

fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 5))

data['price_per_carat'] = data['price'] / data['carat']

cols = ['cut', 'color', 'clarity']
order = [['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'], ['D', 'E', 'F', 'G', 'H', 'I', 'J'], ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']]


for i in cols:
    plt.sca(axes[cols.index(i)])
    sns.boxplot(x = i, y = 'price_per_carat', order = order[cols.index(i)],data=data)
    plt.title(i)
    

plt.show()
```

```python
'''
From the previous plots we see that, in general, data is very disperse. There are a lot of outliners in almost all cases. Also, for cut, higher prices
category is Premium, followed by Very Good. In color, it's G (which also we'll see after it's the most sold), and then H, I. Finally for clarity,
the higher ones are VS2 and VS1.

Considering it as individual parameters you can't see a clear growing pattern for better cut, color or clarity. But this doesn't mean it doesn't affect
to the price.

As seen below, diamonds are not equallt distributed between categories for each parameter (cut, color and clarity). This may also bias the results.
'''

data['cut'].value_counts()
```

```python
data['color'].value_counts()
```

```python
data['clarity'].value_counts()
```

```python
'''
I'm going to represent all values, and see it's distribution per category: cut, color and clarity.
'''

sns.scatterplot(data=data,x='carat',y='price',hue='cut')
```

```python
'''
Again, we see the data is not equally distributed.

In Project 3 I'll see if it affects the model later, but for now, for better visualization, let's try to
make both variables logarithmic.

This way we normalize the data and I'll see the % of price variation per % of carat variation, which is still
valid for what I want to see: if there is any connection between higher cut quality and higher price.
'''

data['log_price'] = np.log(data['price'])
data['log_carat'] = np.log(data['carat'])

plt.figure(figsize = (15,10))
ax = sns.scatterplot(data = data, x = 'log_carat', y = 'log_price', hue = 'cut')
```
